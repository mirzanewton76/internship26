```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
```

```python
X = dataset.iloc[:, :-1].values
```

```python
y = dataset.iloc[:, -1].values
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
```

```python
X_test=sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier()
```

```python
classifier.fit(X_train, y_train)
```

```python
clf2 = DecisionTreeClassifier(
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)
```

```python
clf2.fit(X_train, y_train)
```

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
print(clf2.predict(sc.transform([[51,87000]])))
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred2 = clf2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm2 = confusion_matrix(y_test, y_pred2)
```

```python
print(cm2)
```

```python
accuracy_score(y_test, y_pred2)
```

```python
DecisionBoundaryDisplay.from_estimator(
        classifier,
        X_train,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=dataset.columns[0],
        ylabel=dataset.columns[1],
    )
# Scatter plot
plt.scatter(X_train[:, 0], X_train[:, 1], 
            c=y_train, 
            s=20, edgecolors="k")
plt.title('Decision Tree-Classifier(Training set)')
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
DecisionBoundaryDisplay.from_estimator(
        classifier,
      X_test,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=dataset.columns[0],
        ylabel=dataset.columns[1],
)
plt.scatter(X_test[:, 0], X_test[:, 1], 
            c=y_test, 
            s=20, edgecolors="k")
plt.title('Decision Tree-Classifier(Testing set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
    
```

```python
from sklearn import tree
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_decision_tree(clf_object, feature_names, class_names):
    plt.figure(figsize=(15, 10))
    tree.plot_tree(clf_object, 
                   filled=True, 
                   feature_names=feature_names, 
                   class_names=class_names, 
                   rounded=True)
```

```python
plot_decision_tree(classifier, ['Age','EstimatedSalary'], ['Not Purchased', 'Purchased'])
```

```python
importances = classifier.feature_importances_
print("Feature importances:")
for name, importance in zip(['Age', 'EstimatedSalary'], importances):
    print(f"{name}: {importance:.4f}")
```
